## Task 3 - Data Cleaning and Preparation
From the information obtained above, you will need to fix the data format. 

### Convert to appropriate data type

```python
# Convert to string
covid_ng_data['No. of Cases (Lab Confirmed)'] = covid_ng_data['No. of Cases (Lab Confirmed)'].astype(str)
# Apply string replacement
covid_ng_data['No. of Cases (Lab Confirmed)'] = covid_ng_data['No. of Cases (Lab Confirmed)'].apply(lambda x: x.replace(',', ''))
# Convert to numeric
covid_ng_data['No. of Cases (Lab Confirmed)'] = pd.to_numeric(covid_ng_data['No. of Cases (Lab Confirmed)'])


# Convert to string
covid_ng_data['No. of Cases (on admission)'] = covid_ng_data['No. of Cases (on admission)'].astype(str)
# Apply string replacement
covid_ng_data['No. of Cases (on admission)'] = covid_ng_data['No. of Cases (on admission)'].apply(lambda x: x.replace(',', ''))
# Convert to numeric
covid_ng_data['No. of Cases (on admission)'] = pd.to_numeric(covid_ng_data['No. of Cases (on admission)'])


# Convert to string
covid_ng_data['No. Discharged']  = covid_ng_data['No. Discharged'].astype(str)
# Apply string replacement
covid_ng_data['No. Discharged']  = covid_ng_data['No. Discharged'].apply(lambda x: x.replace(',', ''))
# Convert to numeric
covid_ng_data['No. Discharged']  = pd.to_numeric(covid_ng_data['No. Discharged'] )

```


* From the Real GDP Data, generate a `barplot` using the GDP values for each year & quarters. For example: On x-axis you will have year 2017 and the bars will be values of each quarters(Q1-Q4). You expected to have subplots of each quarters on one graph.
<br>
Hint: Use [Pandas.melt](https://pandas.pydata.org/docs/reference/api/pandas.melt.html) to create your plot DataFrame 
* Set your quarter legend to lower left.
* Using `axhline`, draw a horizontal line through the graph at the value of Q2 2020.
* Write out your observation

```python
# Group the GDP values by Year and Quarter
grouped_data = gdp.groupby(['Year', 'Q1-Q4'])['GDP'].sum().reset_index()

# Filter the data for Q2 2020
q2_2020_data = grouped_data[(grouped_data['Year'] == 2020) & (grouped_data['Q1-Q4'] == 'Q2')]

# Create subplots for each quarter
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
fig, axes = plt.subplots(nrows=1, ncols=len(quarters), figsize=(15, 5), sharey=True)

# Iterate over each quarter and create a bar plot
for i, quarter in enumerate(quarters):
    quarter_data = grouped_data[grouped_data['Q1-Q4'] == quarter]
    axes[i].bar(quarter_data['Year'], quarter_data['GDP'])
    axes[i].set_title(quarter)
    axes[i].set_xlabel('Year')
    axes[i].set_ylabel('GDP')
    
    # Draw a horizontal line at the value of Q2 2020 GDP
    if quarter == 'Q2':
        q2_2020_gdp = q2_2020_data['GDP'].values[0]
        axes[i].axhline(q2_2020_gdp, color='red', linestyle='--')

# Set a common title for the entire plot
fig.suptitle('GDP by Year and Quarter')

# Set the quarter legend to the lower left
fig.legend(quarters, loc='lower left')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```
